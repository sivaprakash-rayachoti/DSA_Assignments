Given an integer¬†`n`, return¬†*`true`¬†if it is a power of two. Otherwise, return¬†`false`*.

An integer¬†`n`¬†is a power of two, if there exists an integer¬†`x`¬†such that¬†`n == 2x`.

**Example 1:**
Input: n = 1 

Output: true

**Example 2:**
Input: n = 16 

Output: true

**Example 3:**
Input: n = 3 

Output: false

</aside>

```python
def isPowerOfTwo(n):
    if n <= 0:
        return False
    return n == n & (n - 1)


print(isPowerOfTwo(1))  # True
print(isPowerOfTwo(16))  # True
print(isPowerOfTwo(3))  # False
```

Given a number n, find the sum of the first natural numbers.

**Example 1:**

Input: n = 3 

Output: 6

**Example 2:**

Input  : 5 

Output : 15

</aside>

```python
def sumOfNaturalNumbers(n):
    return (n * (n + 1)) // 2

# Example usage:
n = 3
sum = sumOfNaturalNumbers(n)
print(sum)  # Output: 6

n = 5
sum = sumOfNaturalNumbers(n)
print(sum)  # Output: 15
```

****Given a positive integer, N. Find the factorial of N. 

**Example 1:**

Input: N = 5 

Output: 120

**Example 2:**

Input: N = 4

Output: 24

</aside>

```python
def factorial(N):
    result = 1
    for i in range(1, N + 1):
        result *= i
    return result


N = 5
print(factorial(N))  # Output: 120

N = 4
print(factorial(N))  # Output: 24
```

üí° **Question 5**

Given an array of integers¬†**arr**, the task is to find maximum element of that array using recursion.

**Example 1:**

Input: arr = {1, 4, 3, -5, -4, 8, 6};
Output: 8

**Example 2:**

Input: arr = {1, 4, 45, 6, 10, -8};
Output: 45

</aside>

```python
def findMax(arr, left, right):
    if left == right:
        return arr[left]

    mid = (left + right) // 2
    maxLeft = findMax(arr, left, mid)
    maxRight = findMax(arr, mid + 1, right)

    return max(maxLeft, maxRight)


arr1 = [1, 4, 3, -5, -4, 8, 6]
print(findMax(arr1, 0, len(arr1) - 1))  # 8

arr2 = [1, 4, 45, 6, 10, -8]
print(findMax(arr2, 0, len(arr2) - 1))  # 45
```

üí° **Question 6**

Given first term (a), common difference (d) and a integer N of the Arithmetic Progression series, the task is to find Nth term of the series.

**Example 1:**

Input : a = 2 d = 1 N = 5
Output : 6
The 5th term of the series is : 6

**Example 2:**

Input : a = 5 d = 2 N = 10
Output : 23
The 10th term of the series is : 23

</aside>

```python
def findNthTerm(a, d, n):
    nth_term = a + (n - 1) * d
    return nth_term

# Example usage:
a = 2
d = 1
n = 5
nth_term = findNthTerm(a, d, n)
print(nth_term)  # Output: 6

a = 5
d = 2
n = 10
nth_term = findNthTerm(a, d, n)
print(nth_term)  # Output: 23
```

üí° **Question 7**

Given a string S, the task is to write a program to print all permutations of a given string.

**Example 1:**

***Input:***

*S = ‚ÄúABC‚Äù*

***Output:***

*‚ÄúABC‚Äù, ‚ÄúACB‚Äù, ‚ÄúBAC‚Äù, ‚ÄúBCA‚Äù, ‚ÄúCBA‚Äù, ‚ÄúCAB‚Äù*

**Example 2:**

***Input:***

*S = ‚ÄúXY‚Äù*

***Output:***

*‚ÄúXY‚Äù, ‚ÄúYX‚Äù*

</aside>

```python
def permutations(S):
    # Convert the string to a list to make it mutable
    S = list(S)
    result = []
    backtrack(S, 0, len(S) - 1, result)
    return result

def backtrack(S, start, end, result):
    if start == end:
        # Add the current permutation to the result
        result.append(''.join(S))
    else:
        for i in range(start, end + 1):
            # Swap the characters
            S[start], S[i] = S[i], S[start]
            
            # Recursively generate permutations for the remaining string
            backtrack(S, start + 1, end, result)
            
            # Restore the original order by swapping back
            S[start], S[i] = S[i], S[start]

            
S = "ABC"
print(permutations(S))  # Output: ['ABC', 'ACB', 'BAC', 'BCA', 'CBA', 'CAB']

S = "XY"
print(permutations(S))  # Output: ['XY', 'YX']

```

üí° **Question 8**

Given an array, find a product of all array elements.

**Example 1:**

Input  : arr[] = {1, 2, 3, 4, 5}
Output : 120
**Example 2:**

Input  : arr[] = {1, 6, 3}
Output : 18

</aside>

```python
def array_product(arr):
    product = 1

    for num in arr:
        product *= num

    return product


# Example 1
arr1 = [1, 2, 3, 4, 5]
result1 = array_product(arr1)
print(result1)  # Output: 120

# Example 2
arr2 = [1, 6, 3]
result2 = array_product(arr2)
print(result2)  # Output: 18
```
